# 2.  How does broadcasting work in NumPy ?

    
  - Broadcasting in NumPy is a way of performing operations on arrays of different shapes without having to manually copy or reshape data.

# Rules of Broadcasting:
If the arrays have different numbers of dimensions, the smaller array is padded with ones on its left side (only virtually — not actually in memory).

If the size in a dimension is different between the arrays, but one of them is 1, NumPy will "stretch" that dimension to match the other array.

If neither dimension is equal or one, broadcasting fails and an error is thrown.

# 6.  What are the differences between NumPy arrays and Python lists ?

#  Data Type:

NumPy arrays: All elements must be of the same type (e.g., all integers or all floats).

Python lists: Can mix types (e.g., integers, strings, objects all together).

# . Speed:

NumPy arrays: Much faster for numerical tasks because they're optimized with C under the hood.

Python lists: Slower because they are more general-purpose.

# Memory Efficiency:

NumPy arrays: Use less memory because they store data in a compact form.

Python lists: Use more memory — each element is a full object with extra metadata.

# . Operations:

NumPy arrays: You can do element-by-element operations directly (like addition, multiplication) without writing loops.

Python lists: You need a loop or list comprehension to do operations on each element.

# Functionality:

NumPy arrays: Come with tons of built-in functions for math, stats, linear algebra, etc.

Python lists: Basic functionality — for complex stuff, you need extra work or other libraries.

#  Flexibility:

NumPy arrays: Best when you have structured, numeric data.

Python lists: Better when you need to store mixed or unpredictable data.

# . Resizing:

NumPy arrays: Fixed size; resizing is expensive and creates a new array.

Python lists: Very easy to grow and shrink — just use .append() or del.

# 10.  What is the significance of hierarchical indexing in Pandas ?

    
  - Hierarchical indexing in Pandas (also known as MultiIndex) allows you to handle and analyze high-dimensional data in a more structured and efficient way. It is useful when working with data that has multiple levels of indexing (e.g., rows and columns that have multiple attributes), making it possible to organize data in a nested, multi-dimensional format.

#  Multiple Index Levels
Hierarchical indexing allows you to create multi-level indices (i.e., a series or dataframe can have more than one index), where each level of the index can represent a different dimension of the data.

For example, you could have a dataset with a hierarchical index that includes both country and year, allowing you to access data at different levels (e.g., country-level or year-level) easily.

# Improved Data Representation
Hierarchical indexing can make complex datasets more intuitive and easier to analyze by grouping related data together at different levels.

It enhances the organization of data, allowing for more readable, efficient data access, manipulation, and aggregation.

#  Efficient Data Selection
You can easily access subsets of data based on the hierarchical index. For example, if you have a MultiIndex with levels for state and city, you can filter the data for a particular state or city without needing to flatten the data structure.

The indexing mechanism provides flexibility to slice data along different dimensions, making complex data filtering much simpler.

# Support for Aggregation and Grouping
Hierarchical indexing allows for more powerful aggregation and group-by operations. With multiple index levels, you can group data by one or more levels of the index and apply aggregation functions (e.g., sum, mean).

It is particularly useful when analyzing time series or grouped data.

#  Better Performance for Complex Operations
Hierarchical indexing optimizes performance when dealing with large datasets that have multiple attributes across rows or columns. It ensures that operations such as merging, reshaping, and grouping are faster and more memory-efficient compared to working with regular indices.

# 15.  How does NumPy handle multidimensional arrays ?

# . ndarray Object:
The core object in NumPy for representing arrays is ndarray. This object is optimized for vectorized operations, which makes it efficient for mathematical operations.

An ndarray can have any number of dimensions, from 0 (scalar) to n-dimensional arrays (e.g., 2D, 3D, etc.).

# Shape :-  
The shape attribute of an ndarray is a tuple that tells you the size of the array along each dimension.

For example:

A 1D array: array([1, 2, 3]) has shape (3,).

A 2D array (matrix): array([[1, 2], [3, 4]]) has shape (2, 2).

A 3D array (tensor): array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]) has shape (2, 2, 2).

#  Indexing:
NumPy supports advanced indexing for multidimensional arrays, allowing you to access individual elements or slices efficiently.

For a 2D array, you can access elements using array[row, column].

For a 3D array, the indexing works as array[depth, row, column].

#  Broadcasting:
NumPy supports broadcasting, which allows you to perform arithmetic operations between arrays of different shapes. The smaller array is "broadcast" across the larger array without needing to duplicate the data.

For instance, adding a scalar to a multidimensional array will add that scalar to every element of the array.

# 21.  Why is NumPy’s array slicing faster than Python’s list slicing ?

# Contiguous Memory Allocation:

NumPy arrays are stored in contiguous blocks of memory, meaning that all elements of the array are placed next to each other. This allows for more efficient access to memory during slicing, as the underlying data is stored in a way that allows for better cache utilization and faster memory access.

Python lists, on the other hand, are arrays of pointers to objects, which makes them less efficient in terms of memory access because they are not contiguous.

# Optimized C Implementation:

NumPy is implemented in C, which allows it to leverage lower-level optimizations and faster memory manipulation. The slicing in NumPy is done in C, which is far faster than Python’s native list slicing, which is interpreted and thus slower.

In contrast, Python lists are high-level objects, and slicing a list involves creating a new list, copying references of the elements from the original list to the new list. This process can be relatively slow.

# No Need for Python Object Wrapping:

In NumPy, elements are stored as fixed-size, homogeneously typed objects (e.g., int32, float64), which means that slicing can directly reference the raw data in memory without needing to wrap the values in Python objects.

Python lists, however, store elements as references to Python objects, which adds overhead in terms of memory access and object management.

# Broadcasting and Advanced Indexing:

NumPy arrays also support broadcasting and advanced indexing mechanisms, which allow for more efficient slicing and manipulation of data across multiple dimensions. This advanced slicing mechanism is highly optimized for performance.

```python
# Q.13 Create a 3D scatter plot using Plotly.
import plotly.graph_objs as go
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)

# Create a 3D scatter plot
scatter = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=5,
        color=z,                # Set color to the z values
        colorscale='Viridis',    # Choose a colorscale
        opacity=0.8
    )
)

layout = go.Layout(
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    margin=dict(l=0, r=0, b=0, t=0)
)

fig = go.Figure(data=[scatter], layout=layout)

# Show the plot
fig.show()
```
